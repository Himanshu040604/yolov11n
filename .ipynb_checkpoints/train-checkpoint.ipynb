```python
import glob, os

base = 'dataset/video1'
splits = ['train', 'valid', 'test']

for split in splits:
    lbl_dir = f'{base}/{split}/labels'
    for path in glob.glob(os.path.join(lbl_dir, '*.txt')):
        lines = []
        with open(path) as f:
            for line in f:
                parts = line.strip().split()
                if not parts:
                    continue
                # only keep original class 0 (cars) and map it to 2
                # if your old labels never use 1, the others will be discarded
                if parts[0] == '0':
                    parts[0] = '2'
                    lines.append(' '.join(parts))
        # overwrite file with only new '2' lines
        with open(path, 'w') as f:
            f.write('\n'.join(lines))
    print(f'→ Processed {split} labels, {len(glob.glob(lbl_dir+"/*.txt"))} files.')
```

```python
import glob

base       = 'dataset/video1'
splits     = ['train', 'valid', 'test']
errors     = []
unique_ids = set()

for split in splits:
    lbl_dir = f'{base}/{split}/labels'
    for path in glob.glob(f'{lbl_dir}/*.txt'):
        with open(path) as f:
            for lineno, line in enumerate(f, start=1):
                parts = line.strip().split()
                if not parts:
                    continue
                cls_id = parts[0]
                unique_ids.add(cls_id)
                if cls_id != '2':
                    errors.append((path, lineno, line.strip()))

print("✔️  Unique class IDs found in labels:", sorted(unique_ids))
if errors:
    print(f"⚠️  Found {len(errors)} erroneous lines (not class '2'):")
    for p, ln, txt in errors[:5]:
        print(f"   {p}, line {ln}: {txt}")
else:
    print("✅ All labels are class 2 only.")
```

```python
# train_cars_cosine.py
from ultralytics import YOLO
import os

# ─── 1. Paths & run settings ───────────────────────────────────────────────────────────
DATA_YAML   = 'dataset/video1/data.yaml'       # your single-class YAML
PRETRAINED  = 'dataset/yolo11n.pt'            # COCO-pretrained weights
PROJECT     = 'outputs'
RUN_NAME    = 'cars_cosine_run'                # output folder name

# ─── 2. Hyperparameters & schedules ────────────────────────────────────────────────────
EPOCHS        = 100
BATCH_SIZE    = 8
IMGSZ         = 640

OPTIMIZER     = 'AdamW'
LR0           = 0.0005      # initial LR
LRF           = 0.01        # final LR = LR0 * LRF (cosine decay)
WARMUP_EPOCHS = 3

WEIGHT_DECAY   = 0.0005
DROPOUT        = 0.2         # dropout in detection head

LABEL_SMOOTHING = 0.1        # optional

PATIENCE     = 10            # early stopping
DEVICE       = 0             # GPU 0
SINGLE_CLS   = True          # treat your one class as idx=0

# ─── 3. Load model & inject dropout ───────────────────────────────────────────────────
model = YOLO(PRETRAINED)
# inject dropout into YOLOv11n's Detect head
model.model.model[-1].m = DROPOUT

# ─── 4. Train (no augmentations) ──────────────────────────────────────────────────────
model.train(
    data            = DATA_YAML,
    epochs          = EPOCHS,
    batch           = BATCH_SIZE,
    imgsz           = IMGSZ,
    optimizer       = OPTIMIZER,
    lr0             = LR0,
    lrf             = LRF,
    warmup_epochs   = WARMUP_EPOCHS,
    weight_decay    = WEIGHT_DECAY,
    dropout         = DROPOUT,
    label_smoothing = LABEL_SMOOTHING,
    device          = DEVICE,
    patience        = PATIENCE,
    single_cls      = SINGLE_CLS,
    project         = PROJECT,
    name            = RUN_NAME,
    save_period     = 0      # only best.pt & last.pt
)

# ─── 5. Prune extra checkpoints ────────────────────────────────────────────────────────
weights_dir = f"{PROJECT}/{RUN_NAME}/weights"
if os.path.isdir(weights_dir):
    for f in os.listdir(weights_dir):
        if f not in ('best.pt', 'last.pt'):
            os.remove(os.path.join(weights_dir, f))
    print(f"✅ Kept only best.pt & last.pt in {weights_dir}")
else:
    print(f"⚠️  Weights directory not found: {weights_dir}")
```

```python
# Make sure figures show up inline in Jupyter
%matplotlib inline

import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO

# Load the trained model
model = YOLO('outputs/cars_cosine_run2/weights/best.pt')

# Path to your test image
image_path = 'dataset/video2/test/images/vid1_mp4-0043_jpg.rf.770459fdc027709484e9124d9e7fbbe4.jpg'

# Run inference
results = model(image_path, conf=0.3)[0]

# Read image with OpenCV
img = cv2.imread(image_path)

# Draw boxes
for box in results.boxes:
    x1, y1, x2, y2 = map(int, box.xyxy[0].cpu().numpy())
    cls  = int(box.cls.cpu().numpy().item())
    conf = float(box.conf.cpu().numpy().item())
    label = f"{model.names[cls]} {conf:.2f}"

    cv2.rectangle(img, (x1, y1), (x2, y2), (0,255,0), 2)
    cv2.putText(img, label, (x1, y1-10),
                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,255,0), 1)

# Convert BGR→RGB and display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,8))
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from ultralytics import YOLO
from IPython.display import clear_output

# 1) Setup paths (adjust to your project layout)
base_dir   = os.getcwd()                             # should be C:/Users/.../yolov11n
frames_dir = os.path.join(base_dir, 'FRAMES')
os.makedirs(frames_dir, exist_ok=True)
video_path = os.path.join(base_dir, 'dataset/test.mp4')              # place your video here
out_path   = os.path.join(base_dir, '/outputs/vidoutput')

# 2) Counters & trackers
entry_count = exit_count = 0
next_id_e = next_id_x = 0
e_tracker = {}
x_tracker = {}
prev_centers = []

# 3) Load your fine-tuned car model
model = YOLO('outputs/cars_cosine_run2/weights/best.pt')  # or your run

# 4) Open video & prepare writer
cap   = cv2.VideoCapture(video_path)
fps   = cap.get(cv2.CAP_PROP_FPS) or 20
w, h  = 1020, 500
fourcc= cv2.VideoWriter_fourcc(*'mp4v')
out   = cv2.VideoWriter(out_path, fourcc, fps, (w, h))

# 5) Define your counting lines/gates
line_y       = 400
entry_gap    = 50;  exit_gap = 90
entry_y      = line_y - entry_gap
exit_y       = line_y + exit_gap

entry_gate = [(550, entry_y), (980 - 80, entry_y)]  # yellow
red_line    = [(550, line_y), (980, line_y)]
blue_line   = [(130, line_y), (500, line_y)]
exit_gate   = [(80 - 90, exit_y), (600 - 90, exit_y)]  # cyan

# 6) Process frames
frame_idx = 0
while True:
    ret, frame = cap.read()
    if not ret: break
    frame = cv2.resize(frame, (w, h))

    # Detection
    results = model(frame, conf=0.3)[0]
    dets = []
    for box in results.boxes:
        if int(box.cls.cpu()) != 0:  # class 0 is 'car' in single_cls mode
            continue
        x1, y1, x2, y2 = box.xyxy[0].cpu().numpy().astype(int)
        cx, cy = (x1 + x2)//2, (y1 + y2)//2
        prev = next((p for p in prev_centers if np.hypot(cx-p[0], cy-p[1])<50), None)
        dets.append({'box':(x1,y1,x2,y2), 'curr':(cx,cy), 'prev':prev})

    # Entry gate logic
    for d in dets:
        cx, cy = d['curr']; pc = d['prev']
        eid = next((i for i,td in e_tracker.items() if np.hypot(cx-td['curr'][0], cy-td['curr'][1])<50), None)
        if eid is None and pc and pc[1]<entry_y<=cy:
            eid = next_id_e; next_id_e += 1
            e_tracker[eid] = {'prev':pc, 'curr':(cx,cy), 'counted':False}
        elif eid is not None:
            e_tracker[eid].update(prev=e_tracker[eid]['curr'], curr=(cx,cy))
        if eid is not None:
            td = e_tracker[eid]
            if not td['counted'] and td['prev'][1]<line_y<=td['curr'][1]:
                entry_count += 1; td['counted']=True

    # Exit gate logic
    for d in dets:
        cx, cy = d['curr']; pc = d['prev']
        xid = next((i for i,td in x_tracker.items() if np.hypot(cx-td['curr'][0], cy-td['curr'][1])<50), None)
        if xid is None and pc and pc[1]>exit_y>=cy:
            xid = next_id_x; next_id_x += 1
            x_tracker[xid] = {'prev':pc, 'curr':(cx,cy), 'counted':False}
        elif xid is not None:
            x_tracker[xid].update(prev=x_tracker[xid]['curr'], curr=(cx,cy))
        if xid is not None:
            td = x_tracker[xid]
            if not td['counted'] and td['prev'][1]>line_y>=td['curr'][1]:
                exit_count += 1; td['counted']=True

    # Draw detections & lines
    for d in dets:
        x1,y1,x2,y2 = d['box']
        cv2.rectangle(frame, (x1,y1), (x2,y2),(0,255,0),2)
    cv2.line(frame, *entry_gate,  (0,255,255),2)
    cv2.line(frame, *red_line,    (0,0,255), 3)
    cv2.line(frame, *blue_line,   (255,0,0), 3)
    cv2.line(frame, *exit_gate,   (255,255,0),2)

    # Overlay counts
    cv2.putText(frame, f"Cars: {len(dets)}",           (60,40),  cv2.FONT_HERSHEY_SIMPLEX,0.6,(0,255,0),2)
    cv2.putText(frame, f"Entry (red): {entry_count}",  (60,70),  cv2.FONT_HERSHEY_SIMPLEX,0.6,(0,0,255),2)
    cv2.putText(frame, f"Exit (blue): {exit_count}",   (60,100), cv2.FONT_HERSHEY_SIMPLEX,0.6,(255,0,0),2)

    # Write out & display
    out.write(frame)
    clear_output(wait=True)
    plt.figure(figsize=(6,4))
    plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

    prev_centers = [d['curr'] for d in dets]
    frame_idx += 1

cap.release()
out.release()
```

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from ultralytics import YOLO
from IPython.display import clear_output

# ─── 0) Setup ─────────────────────────────────────────────────────────────────────────
base_dir    = os.getcwd()                   # e.g. C:/Users/.../yolov11n
frames_dir  = os.path.join(base_dir, 'FRAMES')
os.makedirs(frames_dir, exist_ok=True)

input_video = os.path.join(base_dir, 'dataset/test.mp4')             # your input
output_video = os.path.join(base_dir, 'outputs', 'processed_video.mp4')
os.makedirs(os.path.dirname(output_video), exist_ok=True)

# ─── 1) Load model ───────────────────────────────────────────────────────────────────
model = YOLO('outputs/cars_cosine_run2/weights/best.pt')  # your fine-tuned weights

# ─── 2) Open video & prep writer ────────────────────────────────────────────────────
cap   = cv2.VideoCapture(input_video)
fps   = cap.get(cv2.CAP_PROP_FPS) or 20
w, h  = 1020, 500
fourcc= cv2.VideoWriter_fourcc(*'mp4v')
out   = cv2.VideoWriter(output_video, fourcc, fps, (w, h))

# ─── 3) Precompute your gate lines ───────────────────────────────────────────────────
line_y    = 400
entry_y   = line_y - 50
exit_y    = line_y + 90

entry_gate = [(550, entry_y), (980 - 80, entry_y)]
red_line   = [(550, line_y),     (980, line_y)]
blue_line  = [(130, line_y),     (500, line_y)]
exit_gate  = [(80 - 90, exit_y), (600 - 90, exit_y)]

# ─── 4) Process all frames ───────────────────────────────────────────────────────────
frame_idx = 0
while True:
    ret, frame = cap.read()
    if not ret:
        break
    frame = cv2.resize(frame, (w, h))

    # 4a) Inference
    results = model(frame, conf=0.3)[0]
    dets = []
    for box in results.boxes:
        if int(box.cls.cpu()) != 0:  # single_cls → cars at 0
            continue
        x1, y1, x2, y2 = box.xyxy[0].cpu().numpy().astype(int)
        dets.append((x1, y1, x2, y2))

    # 4b) Draw detections & lines on a copy
    vis = frame.copy()
    for (x1,y1,x2,y2) in dets:
        cv2.rectangle(vis, (x1,y1), (x2,y2), (0,255,0), 2)
    for pts, color in [
        (entry_gate, (0,255,255)), 
        (red_line,   (0,0,255)), 
        (blue_line,  (255,0,0)), 
        (exit_gate,  (255,255,0))
    ]:
        cv2.line(vis, pts[0], pts[1], color, 2)

    cv2.putText(vis, f"Cars: {len(dets)}", (50,50),
                cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255,255,255), 2)

    # 4c) Save frame image
    frame_file = os.path.join(frames_dir, f'frame_{frame_idx:05d}.png')
    cv2.imwrite(frame_file, vis)

    # 4d) Write to output video
    out.write(vis)

    # 4e) (Optional) display live
    clear_output(wait=True)
    plt.figure(figsize=(6,4))
    plt.imshow(cv2.cvtColor(vis, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

    frame_idx += 1

# ─── 5) Release resources ──────────────────────────────────────────────────────────────
cap.release()
out.release()
print(f"✅ Processed {frame_idx} frames.")
print("Frames saved to:", frames_dir)
print("Video saved to:", output_video)
```
